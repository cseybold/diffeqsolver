```python
import numpy as np
import matplotlib.pyplot as plt
import random
from random import sample
```

```python
x1, y1 = random.sample(range(1, 9), 2)
x2, y2 = random.sample(range(1, 9), 2)
```

```python
def system_of_equations(t, vals):
    x, y = vals

    dx_dt = x1*x - y1*y
    dy_dt = x2*x + y2*y

    return [dx_dt, dy_dt]
```

```python
print("Equations:")
print(str(x1) + "x +", str(y1) + "y =", "x'")
print(str(x2) + "x +", str(y2) + "y =", "y'")
```

```python
arr = np.array([[x1, y1], [x2, y2]])
evals, evecs = np.linalg.eig(arr)

print("Eigenvalues:")
print(evals)

print("Eigenvectors:")
print(evecs)
```

```python
print("Solution:")
print("x(t) =", str(round(evecs[0][0], 2)) + "c1*e^" + str(round(evals[0], 2)) + "t +", str(round(evecs[1][0], 2)) + "c2*e^" + str(round(evals[1], 2)) + "t" )
print("y(t) =", str(round(evecs[0][1], 2)) + "c1*e^" + str(round(evals[0], 2)) + "t +", str(round(evecs[1][1], 2)) + "c2*e^" + str(round(evals[1], 2)) + "t" )
```

```python
t = np.linspace(0, 1,1000)
xt = (evecs[0][0])*np.exp((evals[0])*t) + (evecs[1][0])*np.exp((evals[1])*t)
yt = (evecs[0][1])*np.exp((evals[0])*t) + (evecs[1][1])*np.exp((evals[1])*t)
plt.plot(t, xt, 'b', label='x(t)')
plt.plot(t, yt, 'r', label='y(t)')
plt.xlabel('t')
plt.ylabel('Function value')
plt.title('Solution of the System of Differential Equations')
plt.legend()
plt.grid(True)
plt.show()
```
